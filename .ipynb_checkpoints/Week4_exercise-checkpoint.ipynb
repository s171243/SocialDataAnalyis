```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import folium
import datetime

df = pd.read_csv("Police_Department_Incident_Reports__Historical_2003_to_May_2018.csv")
sf_map = folium.Map([37.77919, -122.41914], zoom_start=13)
sf_map
```

```python
folium.Marker([37.77919, -122.41914], popup='City Hall').add_to(sf_map)
sf_map
```

```python
focuscrimes = set(['WEAPON LAWS', 'PROSTITUTION', 'DRIVING UNDER THE INFLUENCE', 'ROBBERY', 'BURGLARY', 'ASSAULT', 'DRUNKENNESS', 'DRUG/NARCOTIC', 'TRESPASS', 'LARCENY/THEFT', 'VANDALISM', 'VEHICLE THEFT', 'STOLEN PROPERTY', 'DISORDERLY CONDUCT'])
df['Date'] = pd.to_datetime(df['Date'])
df["Year"] = df.Date.apply(lambda x: x.year)
df["longitude"] = df.location.apply(lambda x: float(x.replace("POINT (", "").replace(")", "").split()[0]))
df["latitude"] = df.location.apply(lambda x: float(x.replace("POINT (", "").replace(")", "").split()[1]))
assaults = df.loc[(df.Category == "ASSAULT") & (df.Date > datetime.datetime(year=2009, month=1, day=1)) & (df.Date < datetime.datetime(year=2009, month=1, day=5))]
sf_map = folium.Map([37.77919, -122.41914], zoom_start=13)

for _, row in assaults.iterrows():
    #print(row)
    print()
    folium.CircleMarker([row["latitude"], row["longitude"]],
                    radius=2,
                    popup='East London',
                    color='red',
                    ).add_to(sf_map)
    
sf_map
```

```python
from folium.plugins import HeatMap, HeatMapWithTime
sf_map = folium.Map([37.77919, -122.41914], zoom_start=13)
offenses=df.loc[df.Category == "SEX OFFENSES, NON FORCIBLE"]
prostitution=df.loc[df.Category == "PROSTITUTION"]
heat_data = [[row['latitude'],row['longitude']] for index, row in offenses.iterrows()]
HeatMap(heat_data).add_to(sf_map)
sf_map
```

```python
sf_map = folium.Map([37.77919, -122.41914], zoom_start=13)

years = np.sort(df.Year.unique())
heat_data_2 = [[[row['latitude'],row['longitude']] for index, row in prostitution.loc[(df.Year == year)].iterrows()] for year in years]

hm = HeatMapWithTime(heat_data_2,auto_play=True,max_opacity=0.8)
hm.add_to(sf_map)
# Display the map
sf_map
```
